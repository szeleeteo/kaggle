```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.metrics import mean_squared_error
```

```python
x = np.linspace(0, 50, num=51)
y = x + np.random.random(x.shape)*5
```

```python
plt.scatter(x,y);
```

```python
y.mean(), y.std()
```

```python
y2 = x + np.random.random(x.shape)*5
y2.mean(), y2.std()
```

```python
y3 = x + np.random.random(x.shape)*5
y3.mean(), y3.std()
```

```python
y_all = (y + y2 + y3)/3
```

```python
y_all.mean(), y_all.std()
```

```python
y1 = x + 10
y2 = x - 10
```

```python
plt.figure(figsize=(8,5))
plt.plot(x,y, label='original')
plt.plot(x,y1, label='+1')
plt.plot(x,y2, label='-1')
plt.legend()
```

```python
mean_squared_error(y,y1)
```

```python
mean_squared_error(y,y2)
```

```python
y3 = 1.5 * x
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y, label='original')
plt.scatter(x,y3, label='1')
plt.legend()
```

```python
mean_squared_error(y,y3)
```
